# Apriori Algorithm for Market Basket Analysis

This notebook demonstrates the use of the Apriori algorithm for market basket analysis using synthetically generated transaction data that simulates an online grocery store.

```python
import pandas as pd
import numpy as np
from apriori import *
import random
```

## Step 1: Generate Realistic Transaction Data

```python
def generate_realistic_transactions(n_transactions, max_items_per_transaction):
    # Define product categories and items
    product_categories = {
        'Fruits': ['Apple', 'Banana', 'Orange', 'Grapes', 'Strawberries'],
        'Vegetables': ['Carrots', 'Broccoli', 'Tomatoes', 'Spinach', 'Cucumber'],
        'Dairy': ['Milk', 'Cheese', 'Yogurt', 'Butter', 'Cream'],
        'Bakery': ['Bread', 'Croissant', 'Bagel', 'Muffin', 'Cake'],
        'Meat': ['Chicken', 'Beef', 'Pork', 'Salmon', 'Turkey'],
        'Beverages': ['Water', 'Soda', 'Coffee', 'Tea', 'Juice'],
        'Snacks': ['Chips', 'Cookies', 'Popcorn', 'Nuts', 'Chocolate'],
        'Cleaning': ['Detergent', 'Soap', 'Sponges', 'Paper Towels', 'Trash Bags']
    }
    
    all_products = [item for category in product_categories.values() for item in category]
    
    # Create transactions
    transactions = []
    for _ in range(n_transactions):
        n_items = random.randint(1, max_items_per_transaction)
        transaction = random.sample(all_products, n_items)
        
        # Add some common associations
        if 'Bread' in transaction and random.random() < 0.7:
            transaction.append(random.choice(['Butter', 'Cheese']))
        if 'Coffee' in transaction and random.random() < 0.6:
            transaction.append('Milk')
        if 'Chips' in transaction and random.random() < 0.5:
            transaction.append('Soda')
        
        transactions.append(transaction)
    
    return transactions

# Generate synthetic data
n_transactions = 1000
max_items_per_transaction = 10

transactions = generate_realistic_transactions(n_transactions, max_items_per_transaction)
print("Sample transactions:")
for i, transaction in enumerate(transactions[:5]):
    print(f"Transaction {i}: {transaction}")

print(f"\nTotal number of transactions: {len(transactions)}")
print(f"Total number of unique items: {len(set([item for transaction in transactions for item in transaction]))}")
```

## Step 2: Preprocess Data

```python
# Create transaction matrix
transaction_matrix = create_transaction_matrix(transactions)
print("Transaction Matrix:")
print(transaction_matrix.head())

# Get unique items
unique_items = list(set([item for transaction in transactions for item in transaction]))

# Perform one-hot encoding
encoded_df = one_hot_encode(transactions, unique_items)
print("\nEncoded DataFrame:")
print(encoded_df.head())
```

## Step 3: Apply Apriori Algorithm

```python
# Apply Apriori algorithm
min_support = 0.02
min_confidence = 0.3
frequent_itemsets, rules = apply_apriori(transactions, min_support, min_confidence)

print("Frequent Itemsets:")
print(frequent_itemsets)

print("\nAssociation Rules:")
print(rules)
```

## Step 4: Visualize Results

```python
# Visualize association rules
visualize_rules(rules)
```

## Step 5: Interpret Results and Generate Insights

```python
# Print business insights
print_insights(rules)
```
