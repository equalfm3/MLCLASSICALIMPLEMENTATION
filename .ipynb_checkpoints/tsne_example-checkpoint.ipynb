# t-SNE Example

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from tsne import TSNE
```

```python
# Load the digits dataset
digits = load_digits()
X, y = digits.data, digits.target

# Normalize the data
X = (X - np.min(X)) / (np.max(X) - np.min(X))
```

```python
# Apply t-SNE
tsne = TSNE(n_components=2, perplexity=30, learning_rate=200, n_iter=1000, random_state=42)
X_tsne = tsne.fit_transform(X)
```

```python
# Visualize the results
plt.figure(figsize=(10, 8))
scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='tab10')
plt.colorbar(scatter)
plt.title('t-SNE visualization of the digits dataset')
plt.xlabel('t-SNE feature 1')
plt.ylabel('t-SNE feature 2')
plt.show()
```

```python
# Compare with sklearn's t-SNE implementation
from sklearn.manifold import TSNE as SKTSNE

sk_tsne = SKTSNE(n_components=2, perplexity=30, learning_rate=200, n_iter=1000, random_state=42)
X_sk_tsne = sk_tsne.fit_transform(X)

plt.figure(figsize=(16, 6))

plt.subplot(121)
scatter1 = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='tab10')
plt.colorbar(scatter1)
plt.title('Our t-SNE implementation')
plt.xlabel('t-SNE feature 1')
plt.ylabel('t-SNE feature 2')

plt.subplot(122)
scatter2 = plt.scatter(X_sk_tsne[:, 0], X_sk_tsne[:, 1], c=y, cmap='tab10')
plt.colorbar(scatter2)
plt.title('Sklearn t-SNE implementation')
plt.xlabel('t-SNE feature 1')
plt.ylabel('t-SNE feature 2')

plt.tight_layout()
plt.show()
```

```python
# Experiment with different perplexity values
perplexities = [5, 30, 50, 100]
fig, axes = plt.subplots(2, 2, figsize=(16, 16))

for i, perplexity in enumerate(perplexities):
    tsne = TSNE(n_components=2, perplexity=perplexity, learning_rate=200, n_iter=1000, random_state=42)
    X_tsne = tsne.fit_transform(X)
    
    ax = axes[i // 2, i % 2]
    scatter = ax.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='tab10')
    ax.set_title(f't-SNE with perplexity {perplexity}')
    ax.set_xlabel('t-SNE feature 1')
    ax.set_ylabel('t-SNE feature 2')
    plt.colorbar(scatter, ax=ax)

plt.tight_layout()
plt.show()
```
