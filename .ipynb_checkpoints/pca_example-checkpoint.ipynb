```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from pca import PCA

# Generate sample data
X, y = make_blobs(n_samples=300, centers=4, n_features=3, random_state=42)

# Create and fit PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Plot the results
plt.figure(figsize=(12, 5))

# Original data
ax = plt.subplot(121, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y)
ax.set_title("Original Data (3D)")
ax.set_xlabel("Feature 1")
ax.set_ylabel("Feature 2")
ax.set_zlabel("Feature 3")

# PCA-transformed data
plt.subplot(122)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y)
plt.title("PCA-transformed Data (2D)")
plt.xlabel("First Principal Component")
plt.ylabel("Second Principal Component")

plt.tight_layout()
plt.show()

# Print explained variance ratio
total_var = np.sum(np.var(X, axis=0))
explained_var_ratio = np.var(X_pca, axis=0) / total_var
print("Explained variance ratio:", explained_var_ratio)
```
