```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from hierarchical_clustering import HierarchicalClustering
from scipy.cluster.hierarchy import dendrogram
from sklearn.preprocessing import StandardScaler
from matplotlib.patches import Rectangle

# Generate synthetic customer data
np.random.seed(42)
n_customers = 1000

# Create customer features
recency = np.random.randint(1, 365, n_customers)
frequency = np.random.randint(1, 50, n_customers)
monetary = np.random.randint(10, 1000, n_customers)

# Create DataFrame
df = pd.DataFrame({
    'CustomerID': range(1, n_customers + 1),
    'Recency': recency,
    'Frequency': frequency,
    'Monetary': monetary
})

# Normalize the data
scaler = StandardScaler()
X_normalized = scaler.fit_transform(df[['Recency', 'Frequency', 'Monetary']])

# Create and fit the Hierarchical Clustering model
hc = HierarchicalClustering(n_clusters=4, linkage_method='ward')
hc.fit(X_normalized)

# Add cluster labels to the dataframe
df['Cluster'] = hc.labels_

# Plot the dendrogram with zoomed inset
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8), gridspec_kw={'width_ratios': [3, 1]})

# Main dendrogram
dendrogram(hc.linkage_matrix, ax=ax1, leaf_font_size=8, show_leaf_counts=False)
ax1.set_title('Customer Segmentation Dendrogram')
ax1.set_xlabel('Customer Index')
ax1.set_ylabel('Distance')
ax1.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False)

# Zoomed inset
dendrogram(hc.linkage_matrix, ax=ax2, leaf_font_size=8, show_leaf_counts=True,
           truncate_mode='lastp', p=20)  # Show only last 20 merges
ax2.set_title('Zoomed View')
ax2.set_xlabel('Customer Index')
ax2.set_ylabel('Distance')

# Add connecting lines and rectangle
con1 = plt.matplotlib.patches.ConnectionPatch(xyA=(0.9, 0.2), xyB=(0, 0.8), 
                                             coordsA="axes fraction", coordsB="axes fraction",
                                             axesA=ax1, axesB=ax2, color="red")
con2 = plt.matplotlib.patches.ConnectionPatch(xyA=(0.9, 0.8), xyB=(0, 1), 
                                             coordsA="axes fraction", coordsB="axes fraction",
                                             axesA=ax1, axesB=ax2, color="red")
ax1.add_artist(con1)
ax1.add_artist(con2)
rect = Rectangle((0.9, 0.2), 0.1, 0.6, fill=False, color="red", linewidth=2, transform=ax1.transAxes)
ax1.add_patch(rect)

plt.tight_layout()
plt.show()

# Analyze cluster characteristics
cluster_means = df.groupby('Cluster')[['Recency', 'Frequency', 'Monetary']].mean()

# Assign meaningful labels to clusters
def label_cluster(row):
    if row['Recency'] <= 30 and row['Frequency'] > 10 and row['Monetary'] > 500:
        return 'High-Value Active'
    elif row['Recency'] <= 90 and row['Frequency'] > 5 and row['Monetary'] > 250:
        return 'Medium-Value Active'
    elif row['Recency'] > 180 and row['Frequency'] <= 3 and row['Monetary'] <= 100:
        return 'Churned'
    else:
        return 'Low-Value or New'

cluster_means['Segment'] = cluster_means.apply(label_cluster, axis=1)
print("\nCluster Segments:")
print(cluster_means)

# Visualize selected cluster characteristics
selected_clusters = cluster_means.sample(n=2).index  # Randomly select 2 clusters
selected_data = df[df['Cluster'].isin(selected_clusters)]

fig, axes = plt.subplots(1, 3, figsize=(20, 6))
features = ['Recency', 'Frequency', 'Monetary']

for i, feature in enumerate(features):
    for cluster in selected_clusters:
        cluster_data = selected_data[selected_data['Cluster'] == cluster]
        axes[i].hist(cluster_data[feature], bins=20, alpha=0.5, label=f'Cluster {cluster}')
    axes[i].set_title(f'Distribution of {feature}')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Count')
    axes[i].legend()

plt.tight_layout()
plt.show()

# Visualize customer segments (improved pie chart)
segment_sizes = df['Cluster'].map(cluster_means['Segment']).value_counts()
plt.figure(figsize=(12, 8))
wedges, texts, autotexts = plt.pie(segment_sizes, 
                                   labels=segment_sizes.index, 
                                   autopct='%1.1f%%', 
                                   startangle=90, 
                                   pctdistance=0.85,
                                   labeldistance=1.05)

plt.title('Distribution of Customer Segments', y=1.05)
plt.axis('equal')

# Adjust label positions to prevent overlap
for i, autotext in enumerate(autotexts):
    autotext.set_position((autotext.get_position()[0], autotext.get_position()[1] * 0.95))

# Legend
plt.legend(wedges, segment_sizes.index,
           title="Segments",
           loc="center left",
           bbox_to_anchor=(1, 0, 0.5, 1))

plt.tight_layout()
plt.show()

# Business Insights
print("\nBusiness Insights:")
print(f"1. We have identified {len(cluster_means)} distinct customer segments.")
print(f"2. {segment_sizes['High-Value Active']:.1f}% of our customers are high-value active customers.")
print(f"3. {segment_sizes['Churned']:.1f}% of our customers may have churned and need re-engagement strategies.")
print("4. Recommendations:")
print("   - For High-Value Active: Implement loyalty programs and exclusive offers.")
print("   - For Medium-Value Active: Encourage increased purchase frequency with targeted promotions.")
print("   - For Churned: Develop re-engagement campaigns with attractive comeback offers.")
print("   - For Low-Value or New: Focus on onboarding and gradually increasing their engagement.")
```
